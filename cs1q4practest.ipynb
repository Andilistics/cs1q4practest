Average Score Program

```python
print("Welcome, user!")

name = str(input("\nWrite down your name: "))
quiz_scores = []
num_quizzes = 3

for i in range(num_quizzes):
  score = float(input(f"\nWrite down quiz score {i+1}: "))
  quiz_scores.append(score)

print("\nCalculating your results...")
calculate_average_score = sum(quiz_scores) / len(quiz_scores)
print(f"Your average score is {calculate_average_score}.")

if calculate_average_score >= 75:
   print("Status: PASSED")
else:
  print("Status: FAILED")

print(f"\nThank you for using our program, {name}!")
```
